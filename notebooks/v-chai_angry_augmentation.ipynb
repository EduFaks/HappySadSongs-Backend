```python
def found_in_translation(text):
    n = np.random.randint(1,5)
    if n == 1:
        print(ts.google(ts.google(text, from_language='en', to_language='th')))
        return ts.google(ts.google(text, from_language='en', to_language='th'))
    elif n == 2:
        print(ts.google(ts.google(text, from_language='en', to_language='pt')))
        return ts.google(ts.google(text, from_language='en', to_language='pt'))
    elif n == 3:
        print(ts.google(ts.google(text, from_language='en', to_language='es')))
        return ts.google(ts.google(text, from_language='en', to_language='es'))
    else: 
        print(ts.google(ts.google(text, from_language='en', to_language='ja')))
        return ts.google(ts.google(text, from_language='en', to_language='ja'))
```

```python
angry_texts = angry_texts.reset_index(drop=True)
```

```python
angry_texts.head(10)
```

```python
%%time
reverse_translate_angry_1000 = angry_texts.head(1000).apply(found_in_translation)
```

```python
reverse_translate_augment_angry = pd.DataFrame(reverse_translate_angry_1000)
```
